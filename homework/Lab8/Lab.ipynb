```python
from matplotlib import pyplot as plt
import numpy as np
import scipy.signal as sing
li = []
x = []
y = []
with open('Lab8.txt', 'r') as f:
    for str in f:
        li.append(str.split())
```

```python
for i in li:
    x.append(i[0])
    y.append(i[1])
del x[0], y[0]
for i in range(len(x)):
    x[i] = float(x[i])
    y[i] = float(y[i])
```

```python
x_data = [20, 40, 60, 80]
y_data = [1, 2, 3, 4]

fig = plt.figure(figsize = (5, 5))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('time')
ax.set_ylabel('amplitude(V)')
plt.plot(x, y)

plt.show()
```

```python
n = len(y)
Y = np.fft.fft(y)/n
Y = Y[range(int(n/2))]
```

```python
Fs = 100                         # sampling rate
Ts = 1.0/Fs                      # sampling interval

k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
freq = frq[range(int(n/2))]           # one side frequency range
```

```python
plt.plot(freq, abs(Y), 'r-')
plt.xlabel('freq (Hz)')
plt.ylabel('amplitude(V)')
plt.xlim(0.15, 2)

plt.show()
```

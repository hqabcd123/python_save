```python
#Main
import requests
from bs4 import BeautifulSoup
import json
import time
import sqlite3 as lite

base = 'https://lihkg.com/'
uri = 'api_v2/thread/latest?cat_id=1&page={}&count=60&type=now'
url = {}
user = []

res = requests.get(base + uri.format(1))
data = json.loads(res.text)
data
#soup = BeautifulSoup(data, 'lxml')
```

```python
conn = lite.connect(':memory:')
cursor = conn.cursor()
cursor.execute('''CREATE TABLE LIHKG
(quote_post_id text, thread_id text, user_nickname text, post_id text, msg text)''')
conn.commit()
print('create table sussece')
```

```python
#Save thread_id and total page
for scr in data['response']['items']:
    for i in range(scr['total_page']):
        dict = {scr['thread_id']:scr['total_page']}
        url.update(dict)
        #print(scr['title'], ': ', base + 'thread/' + scr['thread_id'] + '/page/' + str(i+1))
    time.sleep(0.1)
```

```python
#Save information of those page
#https://lihkg.com/api_v2/thread/1117179/page/1?order=reply_time
for urls in url:
    for i in range(url[urls]):
        res = requests.get(base + 'api_v2/thread/{}/page/{}?order=reply_time'.format(urls, i+1))
        data = json.loads(res.text)
        for datas in data['response']['item_data']:
            #print(datas)
            dict = {'thread_id':datas['thread_id'], datas['post_id']:{'user_name':datas['user_nickname'], 'msg':datas['msg']}}
            #user.append(dict)
            cursor.execute("INSERT INTO LIHKG VALUES(?, ?, ?, ?, ?)",
                   (datas['quote_post_id'], datas['thread_id'], 
                    datas['user_nickname'], datas['post_id'], datas['msg'].replace('<br />', '')))
conn.commit()
```

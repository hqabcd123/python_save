```python
import requests
from bs4 import BeautifulSoup

base = 'https://www.ptt.cc'
uri = '/bbs/Old-Games/index.html'
res = requests.get(base + uri)
```

```python
soup = BeautifulSoup(res.text)
front_page = ''
for i in soup.select('.action-bar a'):
    try:
        print(base + i['href'], i.text)
        if i.text.find('上頁') != -1:
            front_page = i['href']
            print(front_page)
    except KeyError:
        print(KeyError)
```
